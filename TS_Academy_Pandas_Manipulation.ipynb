# **1. Python Modules and Libraries**
**1.1 What is a Module?**

A module is simply a Python file that contains code such as functions or variables that can be reused in other Python programs.

**You can think of a module as a toolbox:**

instead of rebuilding every tool from scratch, you can import and use tools that already exist.




**Example:** Creating a Module

Create a file called **calculator.py**:

```python
%%writefile calculator.py

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b
```

```python
# main.py
import calculator

result = calculator.add(5, 3)
print(result)  # Output: 8
```

**1.2 What is a Library?**

A library (also called a package) is a collection of related modules that are grouped together to provide a set of tools for a specific purpose.

**A library contains many modules (small toolboxes) that help you do specific tasks.**

**Python has thousands of libraries, but some of the most commonly used ones are:**

* **NumPy** → Used for numerical computing and fast mathematical operations.

* **Pandas** → Used for working with datasets, data tables, and data cleaning.
* **Matplotlib** → Used for creating plots, charts, and visualizations.
* **Scikit-Learn (sklearn)** → Used for machine learning algorithms and model building.

**Example:**

If you want to calculate the square root of a number, instead of writing your own function, you can use the math library

```python
# instead of writing a full function like this
def square_root(number):
    return number ** 0.5

print(square_root(25))
```

```python
# you can just import a math library
import math


print(math.sqrt(25))
```

**Difference Between Modules and Libraries**

| Feature           | **Module**                                                            | **Library**                                         |
| ----------------- | --------------------------------------------------------------------- | --------------------------------------------------- |
| **Definition**    | A single Python file containing code (functions, classes, variables). | A collection of multiple modules packaged together. |
| **Size**          | Small                                                                 | Larger                                              |
| **Purpose**       | Provides specific, small tools                                        | Provides many tools for bigger tasks                |
| **Example Files** | `calculator.py`, `math.py`                                            | pandas, numpy, matplotlib                           |
| **Analogy**       | One small toolbox                                                     | A big toolbox filled with many smaller toolboxes    |

# **6. Summary Statistics**
**df.describe()** is a quick statistical summary tool that gives you important metrics for only the numeric columns in your dataset (unless you specify otherwise).

```python
df.describe()
```

On a DataFrame, Pandas returns the descriptive statistics for each numeric column. Here's what each of those values really means and why they matter in data analysis:

**Count:** Total number of non-null (non-missing) entries in the column.
* Helps you spot missing values

**mean:** The average value of the column: sum of values ÷ number of values.
* helps describe the center or typical behavior of the data

**std:** Standard deviation, which measures the spread or variability in the data.
* A high std means values are widely spread out; a low std means clustered near the mean.

**min:** The lowest value in the column.
* Quickly reveals extreme lows or possible errors.

**25%:** First quartile: 25% of the data lies below this value.
* Useful in understanding the distribution and detecting skewness.

**50%:** Median: 50% of the values are below this (the middle value).
* More robust than the mean in the presence of outliers.

**75%:** Third quartile: 75% of the data lies below this value.
* Helps identify the upper end of “normal” values before outliers.

**max:** The maximum value in the column.
* Shows extreme highs, which could be valid outliers or data entry errors.

  **6.1 Summary Statistics for Categorical Columns (object type)**
  
 When you run df.describe(include='object'), Pandas provides summary information for all non-numeric columns, usually containing text values like brand names, locations, colors, etc.

These columns are typically called categorical variables

```python
df.describe(include='object')
```

**count:** The number of non-null (non-missing) entries in the column.
* Helps you know if there are missing values in a categorical column.

**unique:** Number of distinct categories or different values in that column.
* Helps you understand how diverse or repetitive the column is.

**top:** The most frequently occurring value (mode) in the column.
* Helps identify the dominant category in the data.

**freq:** Number of times the top value appears.
* Shows how much one category dominates the column.


#**7. Filtering Data in Pandas**

Filtering lets you select only the rows that meet certain conditions.
This is helpful when you want to work with a specific part of your dataset.

Why Filtering Matters

Imagine you have 558,837 car records, but you only want to analyze:

* Cars made by Toyota

* Cars priced under $15,000

* Cars from the year 2015

Filtering lets you focus only on the rows you need.

**Basic Filtering (Single Condition)**

The most common way to filter data is using boolean indexing.
You create a True/False condition, and Pandas selects the rows where the condition is True.

Syntax:

df[df['column_name'] condition]


**7.1 Filtering Numeric Columns**

Greater Than (>)

```python
# Cars priced above $50,000
expensive_cars = df[df['sellingprice'] > 50000]
```

Less Than (<)

```python
# Cars priced under $5,000
cheap_cars = df[df['sellingprice'] < 5000]
```

Equal To (==)

```python
# Cars from exactly 2015
cars_2015 = df[df['year'] == 2015]
```



**7.2 Filtering Text (Categorical) Columns**

Exact Match


```python
# All Toyota cars

toyotas = df[df['make'] == 'toyota']
```

```python
# Cars sold in California

california_cars = df[df['state'] == 'ca']
```

**7.3 Multiple Conditions**

Often you need to filter by more than one condition. Use:

* & for AND (both conditions must be true)

* | for OR (at least one condition must be true)

Important: Always wrap each condition in parentheses

**AND Condition (&)**

Both conditions must be true.

```python
# Toyota cars priced under $15,000

cheap_toyotas = df[(df['make'] == 'toyota') & (df['sellingprice'] < 15000)]
```



OR Condition (|)

At least one condition must be true.


```python
# Cars that are either Toyota OR Honda

toyota_or_honda = df[(df['make'] == 'toyota') | (df['make'] == 'honda')]
```

This selects all rows where the make column is either "toyota" or "honda".

# **8. Data Cleaning**

**Cleaning Your Dataset**

Before you can analyze or work with a dataset, you need to find and fix problems in the data.

One of the most common issues in real-world data is missing values — cells that are empty or recorded as NaN (Not a Number).

**Why Cleaning Matters**

* Real-world data is often messy, with problems like:

* Missing values → empty or NaN cells

* Duplicates → repeated rows

* Inconsistencies → typos, wrong formats, or mixed data types

Pandas provides powerful tools to handle these issues so your data is ready for analysis.

```python
df.isna()
```

**This returns a DataFrame of True and False values, where:**

* True means the value is missing

* False means the value is present

To make it more useful, you can check how many missing values exist in each column:

```python
df.isna().sum()
```


**.sum():** Adds up the number of True values (the number of missing values) in each column


**Understanding .isna() vs .isnull()**

Both functions do the same thing — they identify missing values in a DataFrame or Series.

Both return the count of missing values. They are interchangeably used in pandas.

```python
df.isnull().sum()
```

#**9. Dropping Null Values in a Dataset (dropna())**
In real-world datasets, it's common to have missing values, empty cells or entries marked as NaN (Not a Number). These can occur due to:

* Incomplete data collection
* Data entry errors
* Data corruption
* Not applicable fields

Dropping nulls is one way to clean your dataset, especially when:

* The number of missing values is small
* The missing rows are not critical to your analysis
* You want to ensure smooth modeling or analysis (most ML algorithms don't handle nulls well)

Caution: Dropping rows with nulls can lead to data loss and potential bias if the missing data follows a pattern. Consider alternatives like imputation (filling missing values) or using algorithms that handle missing data natively.

```python
# Calculate percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100  #/len(df)Divides the null count by the total number of rows to get a proportion
print (missing_percentage)
```

```python
# Drop all rows that have at least one missing value
df_clean = df.dropna()
```

* Removes every row that has at least one null value

* The result is stored in a new DataFrame df_clean

```python
print("Before:", df.shape)
print("After :", df_clean.shape)
```

**9.1 Dropping Columns That Contain Any Missing Values**

# **11. Renaming Columns in Pandas**

Renaming columns is a common task when working with datasets. You might need to rename columns to:

* Make names more readable or descriptive
* Clean up messy or inconsistent names
* Standardize naming conventions (lowercase, no spaces)
* Match column names with another dataset for merging
* Make names Python-friendly (no spaces or special characters)

**11.1 Renaming Specific Columns**

**Using rename() Method**

The most flexible way to rename columns. Pass a dictionary mapping old names to new names.

**Syntax:**

df.rename(columns={'old_name': 'new_name'}, inplace=True)
# OR
df = df.rename(columns={'old_name': 'new_name'})

**11.2 Single Column:**

```python
# Rename 'sellingprice' to 'selling_price'
df = df.rename(columns={'sellingprice': 'selling_price'})

print(df.columns.tolist())
```

**11.3 Multiple Columns:**

```python
# Rename multiple columns at once

df = df.rename(columns={
    'sellingprice': 'selling_price',
    'mmr': 'market_value',
    'vin': 'vehicle_id',
    'saledate': 'sale_date'
})

print(df.columns.tolist())
```

**11.4 Using inplace=True:**

Note: **inplace=True** modifies the original DataFrame. Without it, you must assign the result back to df.

```python
# Modify the original DataFrame directly

df.rename(columns={'odometer': 'mileage'}, inplace=True)
```



It tells pandas:

    Change the column name from 'odometer' to 'mileage'.

inplace=True

This means:

    Apply the change directly to df instead of returning a new DataFrame.

# **12. GroupBy and Aggregations in Pandas**

The GroupBy operation in Pandas is used to analyze data by categories and compute summary statistics for each group. This is essential when you want to understand patterns or compare groups within your dataset.

**With groupby(), you can answer practical questions such as:**

* Which car brand has the highest average selling price?

* How many cars were sold in each state?

* What is the average condition rating for each seller?


**12.1 How GroupBy Works**

You split your DataFrame into groups based on the values in one or more columns, and then apply an aggregation function.

Common aggregation functions include:

* .mean() — average

* .sum() — total

* .count() — number of rows

* .min() — smallest value

* .max() — largest value

* .median() — middle value

You can group by one column or multiple columns, and then chain the result with sorting, filtering, or plotting.

1. Average selling price per car brand

```python
df.groupby('make')['selling_price'].mean()
```

gives a simple, clean summary of average price per brand — way easier to interpret than scanning all 500,000+ rows.

2. Number of cars sold in each state

```python
df.groupby('state')['selling_price'].count().sort_values(ascending=False).head(10)
```

Shows the top-selling states by average price

**12.2 Grouping by Multiple Columns**

```python
df.groupby(['make', 'state'])['selling_price'].mean()
```

* It helps you see patterns, trends, and insights quickly.

* Our brains understand pictures faster than tables.

* A well-designed chart can tell a story at a glance that might take pages of numbers to explain.

| Chart Type         | When to Use                      | Code Example                    |
| ------------------ | -------------------------------- | ------------------------------- |
| **Line**           | Trends over time                 | `plt.plot(x, y)`                |
| **Bar**            | Compare categories               | `plt.bar(x, y)`                 |
| **Horizontal Bar** | Long category names              | `plt.barh(x, y)`                |
| **Histogram**      | Data distribution                | `plt.hist(data)`                |
| **Pie**            | Parts of a whole                 | `plt.pie(values, labels=names)` |
| **Scatter**        | Relationship between 2 variables | `plt.scatter(x, y)`             |
| **Box Plot**       | Distribution & outliers          | `plt.boxplot(data)`             |


**13.2 Basic Structure of a Matplotlib Chart**

Every Matplotlib chart follows 4 simple steps:

```python
import matplotlib.pyplot as plt
```

```python
# 1. Create the plot (figure size)
plt.figure(figsize=(10, 6))  # Width x Height in inches

# 2. Add your data
plt.plot(x_data, y_data)      # Or use plt.bar(), plt.scatter(), etc.

# 3. Customize the chart
plt.title("My Chart Title")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# 4. Display the chart
plt.show()
```

3. Customize it → Add title, axis labels, and other formatting.

3. Display it → plt.show() renders the chart on screen.

1. **Line Plot – Best for Showing how values change over time**

Year vs average selling price.

```python
yearly = df.groupby('year')['selling_price'].mean()

plt.plot(yearly.index, yearly.values)
plt.xlabel('Year')
plt.ylabel('Average Selling Price')
plt.title('Average Car Price Over the Years')
plt.show()
```

2. **Bar Chart – Best for Comparing values across different categories**

top 5 car brands by count.

```python
top_makes = df['make'].value_counts().head()

plt.bar(top_makes.index, top_makes.values)
plt.xticks(rotation=45)
plt.xlabel('Make')
plt.ylabel('Count')
plt.title('Top 5 Car Brands')
plt.show()
```

3. **Horizontal Bar Chart – For Long Category Names**

top 10 models by count.

```python
top_models = df['model'].value_counts().head(10)

plt.barh(top_models.index, top_models.values)
plt.xlabel('Count')
plt.ylabel('Model')
plt.title('Top 10 Car Models')
plt.show()
```

4. **Histograms - Distribution of Values**

Best for: Understanding how values are distributed

```python
plt.hist(df['selling_price'], bins=30)
plt.xlabel('selling_price')
plt.ylabel('Frequency')
plt.title('selling_price Distribution')
plt.show()
```

5. **Pie Charts - Parts of a Whole**

Best for: Showing proportions

```python
colors = df['color'].value_counts().head(5)

plt.pie(colors.values, labels=colors.index, autopct='%1.1f%%')
plt.title('Top 5 Car Colors')
plt.show()
```

6. **Scatter Plots - Relationships Between Variables**

Best for: Showing correlation between two variables

```python
plt.scatter(df['market_value'], df['selling_price'], alpha=0.3)
plt.xlabel('market_value')
plt.ylabel('Selling Price')
plt.title('market_value vs Selling Price')
plt.show()
```

7. **Box Plots - Statistical Distribution**

Best for: Comparing distributions and identifying outliers

```python
plt.boxplot(df['selling_price'].dropna())
plt.ylabel('Selling Price')
plt.title('Selling Price Distribution')
plt.show()
```
